Multivarient Analysis - 

```python
rest = sns.load_dataset('tips')
rest.head(5)
```

```python
flight = sns.load_dataset('flights')
flight.head(5)
```

```python
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
#scatter plot - Numerical - Numerical
sns.scatterplot(x='total_bill', y='tip', data=rest,hue=rest['sex'],style=rest['smoker'],size=rest['size'])
```

```python
#Bar plot - Numerical data and categorical 
sns.barplot(x = df['Education'],y = df['ApplicantIncome'],hue = df['Self_Employed'])
```

```python
#boxplot - Numerical to categorical
sns.boxplot(x = df['Education'],y = df['ApplicantIncome'])
```

```python
#Displot - Numerical to categorical
sns.distplot(df['Education'],df['ApplicantIncome'])
```

```python
#Heatmap  - Categorical to Categorical
sns.heatmap(pd.crosstab(df['Education'],df['Gender']))
```

```python
#clusterMap - categorical to Categorical - tells the closeness of the items
sns.clustermap(pd.crosstab(df['Education'],df['Gender']))
```

```python
#pairplot  - it detects all the numerical columns and then gives the relation btw them in scattered plot
sns.pairplot(iris,hue = 'species')
```
